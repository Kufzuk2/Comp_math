```python
import numpy as np
import matplotlib.pyplot as plt

#  5 исходных функций
def func1(x):
    return np.sin(x * x)

def func2(x):
    return np.cos(np.sin(x))

def func3(x):
    return np.exp(np.sin(np.cos(x)))

def func4(x):
    return np.log(x + 3)

def func5(x):
    return np.power(x + 3, 0.5)


#  5 исходных производных
def func1_der(x):
    return 2 * x * np.cos(x * x)

def func2_der(x):
    return -np.sin(np.sin(x)) * np.cos(x)

def func3_der(x):
    return -func3(x) * np.cos(np.cos(x)) * np.sin(x)

def func4_der(x):
    return 1 / (x + 3)
def func5_der(x):
    return 1 / (2 * func5(x))


# 5 способов вычисления производной
def der1(f, x, h):
    return (f(x + h) - f(x)) / h

def der2(f, x, h):
    return (f(x) - f(x - h)) / h

def der3(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

def der4(f, x, h):
    return 4 * (f(x + h) - f(x - h)) / (6 * h) - (f(x + 2 * h) - f(x - 2 * h)) / (12 * h)

def der5(f, x, h):
    return 3 * (f(x + h) - f(x - h)) / (4 * h) - 3 * (f(x + 2 * h) - f(x - 2 * h)) / (20 * h) + (f(x + 3 * h) - f(x - 3 * h)) / (60 * h)




#точка
x_value = 1

funcs     = [    func1,     func2,     func3,     func4,     func5]
ders      = [     der1,      der2,      der3,      der4,      der5]
true_ders = [func1_der, func2_der, func3_der, func4_der, func5_der]

func_names = ["sin(x^2)", "cos(sin(x))", "exp(cos(sin(x)))", "log(x + 3)", "(x + 3)^{0.5}"]

for i in range(5):
  x = []
  y_1 = []
  y_2 = []
  y_3 = []
  y_4 = []
  y_5 = []

  for n in range(1, 20):
    h = 2 / (2 ** n)
    x.append(h)
    y_1.append(abs(der_1(funcs[i], x_value, h) - true_ders[i](x_value)))
    y_2.append(abs(der_2(funcs[i], x_value, h) - true_ders[i](x_value)))
    y_3.append(abs(der_3(funcs[i], x_value, h) - true_ders[i](x_value)))
    y_4.append(abs(der_4(funcs[i], x_value, h) - true_ders[i](x_value)))
    y_5.append(abs(der_5(funcs[i], x_value, h) - true_ders[i](x_value)))

  plt.plot(x, y_1, label=f"$f'(x) = \sin(x^2)$", marker='o')
  plt.plot(x, y_2, label=f"$f'(x) = \sin(x^2)$", marker='o')
  plt.plot(x, y_3, label=f"$f'(x) = \sin(x^2)$", marker='o')
  plt.plot(x, y_4, label=f"$f'(x) = \sin(x^2)$", marker='o')
  plt.plot(x, y_5, label=f"$f'(x) = \sin(x^2)$", marker='o')
  plt.xscale("log")
  plt.yscale("log")
  plt.xlabel("$h$ ")
  plt.ylabel("$\\Delta f'$ - погрешность производной")
  plt.title(f"График погрешностей для производной функции ${func_names[i]}$")
  #plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2))
  plt.show()
```
